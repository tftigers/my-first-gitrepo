```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

time= np.arange(10)
height= np.array([450,445,430,409,375,331,280,215,144,59])
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

time = np.arange(10)
height = np.array([450,445,430,409,375,331,280,215,144,59])

coef= np.polyfit(time, height, 2)
equ = np.poly1d(coef)

x_plot = np.linspace(0,10,100)
y_plot = equ(x_plot)
plt.plot(x_plot, y_plot, color= 'g')

plt.scatter(time,height,label= 'Height of a ball')
plt.title("Time vs Height")
plt.xlabel('Time')
plt.ylabel('Height')
plt.legend()
plt.show()
```
